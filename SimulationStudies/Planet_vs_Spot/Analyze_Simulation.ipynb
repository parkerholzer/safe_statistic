```python
Dopp_pvals = pd.read_csv("Dopp_Pvals_sn150_2.csv", header=None)
Spot_pvals = pd.read_csv("../simulation3/activityindicators/Spot_Pvals_sn150_2.csv", header=None)
```

```python
falservs = pd.read_csv("SOAP_RVs.csv").RV.values
```

```python
fig, (ax1, ax2) = plt.subplots(2,1, sharex=True, figsize=(9,6))
#ax1.set_xlabel("Relative Time", fontsize=12)
ax1.set_ylabel("P-values", fontsize=12)
ax1.boxplot(np.array(Dopp_pvals))
ax1.set_title("10 m/s Exoplanet", fontsize=14)

ax1p = ax1.twinx()
ax1p.set_ylabel("Detected Radial Velocity (m/s)", fontsize=12, color='red')
ax1p.plot(np.arange(1,26), 10*np.sin(2*np.pi*np.arange(1,26)/25), color='red')
ax1p.tick_params(axis='y', labelcolor='red', color='red')

ax2.boxplot(np.array(Spot_pvals))
ax2.set_ylabel("P-values", fontsize=12)
ax2.set_xlabel("Relative Time", fontsize=12)

ax2p = ax2.twinx()
ax2p.set_ylabel("Detected Radial Velocity (m/s)", fontsize=12, color='red')
ax2p.plot(np.arange(1,26), falservs, color='red')
ax2p.tick_params(axis='y', labelcolor='red', color='red')
ax2p.set_title("1% Equatorial Spot", fontsize=14)

fig.tight_layout()

plt.savefig("exoplanet_vs_spot_simulation")
plt.show()
```
