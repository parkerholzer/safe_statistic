```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sys
sys.path.append("../")
from abfeature_functions import lowerSNR
import glob
```

# Simulate the data

```python
filenames = glob.glob("../../Methods/SOAP/integr*")
filenames = np.array(filenames)[np.argsort(np.array([float(f.split("phase_")[1].split(".csv")[0]) for f in filenames]))]
spec = [pd.read_csv(f) for f in filenames]
```

```python
tmp = pd.read_csv("../../Methods/SOAP/integrated_spectrum_full_reso_spot_prot_25.0_size_0.010_lat_0_phase_-0.5000.csv")
tmp["Flux"] = tmp.Flux.values/np.max(tmp.Flux.values)
keep = np.where((tmp.Flux.values > 0) & (tmp.Wavelength.values > 4500))[0]
outputspec = []
for s in spec:
    unc = 1/np.sqrt(tmp.Flux.values[keep])
    flx = lowerSNR(s.Flux.values[keep]/np.max(s.Flux.values[keep]), 500)
    wvl = s.Wavelength.values[keep]
    outputspec.append(pd.DataFrame({"Wavelength": wvl, "Flux": flx, "Uncertainty": unc}))
```

```python
for t in range(len(spec)):
    outputspec[t].to_csv("soap_1pcnt_snr500_%d.csv"%t)
```
