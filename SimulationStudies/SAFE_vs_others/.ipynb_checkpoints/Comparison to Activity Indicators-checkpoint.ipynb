```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib notebook
```

```python
bissnr = [100, 110, 120, 130, 140, 150, 160, 170, 180, 190, 200, 250, 300, 350, 400, 450, 500]
bis = [pd.read_csv("activityindicators/191029_snr%d_BIS.csv"%sn, header=None) for sn in bissnr]
bislopesnr = [100, 110, 120, 130, 140, 150, 160, 170, 180, 190, 200, 250, 300, 350, 400, 450, 500]
bislope = [pd.read_csv("activityindicators/191029_snr%d_BISlope.csv"%sn, header=None) for sn in bislopesnr]
bigausssnr = [100, 110, 120, 130, 140, 150, 160, 170, 180, 190, 200, 250, 300, 350, 400, 450, 500]
bigauss = [pd.read_csv("activityindicators/191029_snr%d_biGauss.csv"%sn, header=None) for sn in bigausssnr]
vspansnr = [100, 110, 120, 130, 140, 150, 160, 170, 180, 190, 200, 250, 300, 350, 400, 450, 500]
vspan = [pd.read_csv("activityindicators/191029_snr%d_Vspan.csv"%sn, header=None) for sn in vspansnr]
fwhmsnr = [100, 110, 120, 130, 140, 150, 160, 170, 180, 190, 200, 250, 300, 350, 400, 450, 500]
fwhm = [pd.read_csv("activityindicators/191029_snr%d_fwhm.csv"%sn, header=None) for sn in fwhmsnr]
```

```python
nullbis = np.hstack((np.array(bis[5].iloc[:,:7]), np.array(bis[5].iloc[:,-6:]))).reshape((1000*13, 1))
crit1 = np.percentile(nullbis, 0.5)
crit2 = np.percentile(nullbis, 99.5)
phs = 16
print(len(np.where((bis[5].iloc[:,phs] < crit1) | (bis[5].iloc[:,phs] > crit2) )[0])/1000)
plt.hist(nullbis, bins=50, density=True)
plt.hist(bis[5].iloc[:,phs], histtype='step', bins=40, density=True)

plt.show()
```

```python
crit1 = []
crit2 = []
for i in range(11):
    nullbis = np.hstack((np.array(bis[i].iloc[:,:7]), np.array(bis[i].iloc[:,-6:]))).reshape((1000*13, 1))
    crit1.append(np.percentile(nullbis, 0.5))
    crit2.append(np.percentile(nullbis, 99.5))

power_bis = [[len(np.where((bis[i].iloc[:,j] < crit1[i]) | (bis[i].iloc[:,j] > crit2[i]) )[0])/1000 for j in range(7,19)]
            for i in range(1,11)]

plt.rc('text', usetex=True)
plt.imshow(power_bis, cmap='Purples', origin='lower')
plt.yticks(np.arange(10), ["%d"%sn for sn in bissnr[1:11]])
plt.xticks(np.arange(12), [r"%d$^{\circ}$"%int(k) for k in np.arange(7,19)*360/25])
plt.xlabel("Phase", fontsize=14)
plt.ylabel("SNR", fontsize=14)
plt.tick_params(axis='y', labelsize=14)
plt.tick_params(axis='x', labelsize=12)
cbar = plt.colorbar()
cbar.set_label("Power", fontsize=14)
cbar.ax.tick_params(labelsize=14)
plt.clim(0,1)
plt.tight_layout()
plt.show()
```

```python
crit1 = 0.01*np.ones(10)

power_safe = [[len(np.where(safe[i].iloc[:,j] < crit1[i] )[0])/1000 for j in range(7,19)]
            for i in range(10)]

plt.rc('text', usetex=True)
plt.imshow(np.array(power_safe) - np.array(power_fwhm), cmap='PuOr', origin='lower')
plt.yticks(np.arange(10), ["%d"%sn for sn in snr])
plt.xticks(np.arange(12), [r"%d$^{\circ}$"%int(k) for k in np.arange(7,19)*360/25])
plt.xlabel("Phase", fontsize=14)
plt.ylabel("SNR", fontsize=14)
plt.tick_params(axis='y', labelsize=14)
plt.tick_params(axis='x', labelsize=12)
cbar = plt.colorbar()
cbar.set_label("SAFE Power - FWHM Power", fontsize=14)
cbar.ax.tick_params(labelsize=14)
mx = np.max(np.absolute(np.array(power_safe) - np.array(power_fwhm)))
plt.clim(-mx, mx)
plt.tight_layout()
plt.savefig("SAFE-FWHM_power.pdf", format='pdf')
plt.show()
```

```python
fig, ax = plt.subplots(2,3, figsize=(9,7))
plt.rc('text', usetex=True)
images= []

images.append(ax[0,0].imshow(power_bis, cmap='Purples', origin='lower'))
ax[0,0].set_yticks(np.arange(0,10,2))
ax[0,0].set_yticklabels(["%d"%sn for sn in np.array(bissnr)[np.arange(1,10,2)]], fontsize=14)
ax[0,0].set_xticks(np.array([0,5.5,11]))
ax[0,0].set_xticklabels([""]*3)
ax[0,0].set_title("BIS", pad=0.2, fontsize=15)
ax[0,0].set_ylabel("SNR", fontsize=14)

images.append(ax[0,1].imshow(power_bislope, cmap='Purples', origin='lower'))
ax[0,1].set_yticks(np.arange(0,10,2))
ax[0,1].set_yticklabels([""]*5)
ax[0,1].set_xticks(np.array([0,5.5,11]))
ax[0,1].set_xticklabels([""]*3)
ax[0,1].set_title("BiSlope", pad=0.2, fontsize=15)

images.append(ax[0,2].imshow(power_bigauss, cmap='Purples', origin='lower'))
ax[0,2].set_yticks(np.arange(0,10,2))
ax[0,2].set_yticklabels([""]*5)
ax[0,2].set_xticks(np.array([0,5.5,11]))
ax[0,2].set_xticklabels([""]*3)
ax[0,2].set_title("BiGauss", pad=0.2, fontsize=15)

images.append(ax[1,0].imshow(power_vspan, cmap='Purples', origin='lower'))
ax[1,0].set_yticks(np.arange(0,10,2))
ax[1,0].set_yticklabels(["%d"%sn for sn in np.array(vspansnr)[np.arange(1,10,2)]], fontsize=14)
ax[1,0].set_xticks(np.array([0,5.5,11]))
ax[1,0].set_xticklabels([r"%d$^{\circ}$"%int(k) for k in np.array([100,180,260])], fontsize=14)
ax[1,0].set_ylabel("SNR", fontsize=14)
ax[1,0].set_xlabel("Phase", fontsize=14)
ax[1,0].set_title("V-span", pad=0.2, fontsize=15)

images.append(ax[1,1].imshow(power_fwhm, cmap='Purples', origin='lower'))
ax[1,1].set_yticks(np.arange(0,10,2))
ax[1,1].set_yticklabels([""]*5)
ax[1,1].set_xticks(np.array([0,5.5,11]))
ax[1,1].set_xticklabels([r"%d$^{\circ}$"%int(k) for k in np.array([100,180,260])], fontsize=14)
ax[1,1].set_xlabel("Phase", fontsize=14)
ax[1,1].set_title("FWHM", pad=0.2, fontsize=15)

images.append(ax[1,2].imshow(power_safe, cmap='Purples', origin='lower'))
ax[1,2].set_yticks(np.arange(0,10,2))
ax[1,2].set_yticklabels([""]*5)
ax[1,2].set_xticks(np.array([0,5.5,11]))
ax[1,2].set_xticklabels([r"%d$^{\circ}$"%int(k) for k in np.array([100,180,260])], fontsize=14)
ax[1,2].set_xlabel("Phase", fontsize=14)
ax[1,2].set_title("SAFE", pad=0.2, fontsize=15)

norm = colors.Normalize(vmin=0, vmax=1)
for im in images:
    im.set_norm(norm)
    
cbar = fig.colorbar(images[-1], ax=ax, pad = 0.15, orientation='horizontal')

cbar.set_label("Power", fontsize=16)
cbar.ax.tick_params(labelsize=14)

plt.savefig("StellarActivity_Comp.pdf", format='pdf')
plt.show()
```
