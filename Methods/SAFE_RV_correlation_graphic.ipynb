```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib notebook
```

```python
theta = np.linspace(0, 2*np.pi, 1000)
backtheta = np.linspace(0, np.pi, 500)
fronttheta = np.linspace(np.pi, 2*np.pi, 500)
fig, ax = plt.subplots(1,2, figsize=(8,4.7))
plt.rc('text', usetex=False)
ax[0].plot(np.zeros(2), [-1.2,1.2], lw=1, c='k', ls='--')
ax[0].plot(np.cos(theta), np.sin(theta), lw=2, c='blue')
ax[0].plot(0.2*np.cos(theta)*np.cos(np.arcsin(0.4)) + 0.4, 0.2*np.sin(theta) + 0.1*np.sin(np.arccos(0.4)), c='m', ls=':')
ax[0].plot(np.cos(backtheta), 0.1*np.sin(backtheta), c='blue', ls=':')
ax[0].plot(np.cos(fronttheta), 0.1*np.sin(fronttheta), c='blue')
ax[0].plot(np.cos(fronttheta)*np.cos(np.arcsin(0.6)), 0.1*np.sin(fronttheta) + 0.6, c='blue')
ax[0].scatter([np.cos(13*np.pi/8)*np.cos(np.arcsin(0.6)),np.cos(11*np.pi/8)*np.cos(np.arcsin(0.6))], 
              [0.1*np.sin(13*np.pi/8) + 0.6,0.1*np.sin(11*np.pi/8) + 0.6], c='blue', marker='>', s=65)
ax[0].plot(np.cos(backtheta)*np.cos(np.arcsin(0.6)), 0.1*np.sin(backtheta) + 0.6, c='blue', ls=':')
ax[0].plot(np.cos(fronttheta)*np.cos(np.arcsin(0.6)), 0.1*np.sin(fronttheta) - 0.6, c='blue')
ax[0].scatter([np.cos(13*np.pi/8)*np.cos(np.arcsin(0.6)), np.cos(11*np.pi/8)*np.cos(np.arcsin(0.6))],
             [0.1*np.sin(13*np.pi/8) - 0.6,0.1*np.sin(11*np.pi/8) - 0.6], c='blue', marker='>', s=65)
ax[0].plot(np.cos(backtheta)*np.cos(np.arcsin(0.6)), 0.1*np.sin(backtheta) - 0.6, c='blue', ls=':')
ax[0].plot(0.2*np.cos(theta), 0.2*np.sin(theta) - 0.1, c='m')
ax[0].plot(0.2*np.cos(theta)*np.cos(np.arcsin(0.7)) - 0.7, 0.2*np.sin(theta) - 0.1*np.sin(np.arccos(-0.7)), c='m')
ax[0].plot(0.2*np.cos(theta)*np.cos(np.arcsin(0.7)) + 0.7, 0.2*np.sin(theta) - 0.1*np.sin(np.arccos(-0.7)), c='m')
ax[0].text(0 - 0.08,-0.1 - 0.08, "2", fontsize=25, fontweight='bold')
ax[0].text(-0.7 - 0.08,- 0.1*np.sin(np.arccos(-0.7)) - 0.08, "1", fontsize=25, fontweight='bold')
ax[0].text(0.7 - 0.08, -0.1*np.sin(np.arccos(0.7)) - 0.08, "3", fontsize=25, fontweight='bold')
ax[0].text(0.4 - 0.08, +0.1*np.sin(np.arccos(0.4)) - 0.08, "4", fontsize=25, fontweight='bold')
ax[0].spines['top'].set_visible(False)
ax[0].spines['right'].set_visible(False)
ax[0].spines['bottom'].set_visible(False)
ax[0].spines['left'].set_visible(False)
ax[0].set_xticks([])
ax[0].set_yticks([])

ax[1].plot(np.cos(theta), np.sin(theta) + 1, lw=2, c='r', ls='--')
ax[1].plot(1.1*np.cos(theta), 1.1*np.sin(theta) + 1, lw=2, c='w', ls='--')
ax[1].text(1 - 0.08, 1 - 0.08, "1", fontsize=25, fontweight='bold', 
           bbox={'facecolor': 'white', 'alpha': 1.0, 'pad': 3})
ax[1].text(0 - 0.08, 2 - 0.08, "2", fontsize=25, fontweight='bold', 
           bbox={'facecolor': 'white', 'alpha': 1.0, 'pad': 3})
ax[1].text(-1 - 0.08, 1 - 0.08, "3", fontsize=25, fontweight='bold', 
           bbox={'facecolor': 'white', 'alpha': 1.0, 'pad': 3})
ax[1].text(0 - 0.08, 0 - 0.08, "4", fontsize=25, fontweight='bold', 
           bbox={'facecolor': 'white', 'alpha': 1.0, 'pad': 3})
ax[1].set_xticks([-1,0,1])
ax[1].set_xticklabels(["-", "0.0", "+"], fontsize=14)
ax[1].yaxis.set_label_position("right")
ax[1].yaxis.set_ticks_position("right")
ax[1].set_yticks([0,1,2])
ax[1].set_yticklabels(["0.0", "", "High"], fontsize=14)
plt.rc('text', usetex=True)
ax[1].set_xlabel(r"RV (m s$^{-1}$)", fontsize=16)
plt.rc('text', usetex=False)
ax[1].set_ylabel("SAFE", fontsize=16)

plt.tight_layout()
plt.savefig("RV_SAFE_corrgraphic.pdf", format='pdf')
plt.show()
```
