```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib notebook
```

```python
soap1 = pd.read_csv("SOAP_Features.csv")
soap = soap1[(soap1.SOAPamp_0.values > -0.8) &
             (soap1.Continuum.values < 1.02) &
             (soap1.Continuum.values > 0.92)]
```

```python
soap.columns
```

```python
plt.hist(soap.Continuum.values)
plt.show()
```

```python
plt.scatter(soap.MinWvl.values, soap.Continuum.values)
plt.show()
```

```python
fig, ((ax0, ax2, ax3), (ax4, ax5, ax6)) = plt.subplots(nrows = 2, ncols = 3, figsize=(8,5))

ax0.yaxis.set_label_position("right")
ax0.yaxis.tick_right()
ax2.yaxis.set_label_position("right")
ax2.yaxis.tick_right()
ax3.yaxis.set_label_position("right")
ax3.yaxis.tick_right()
ax4.yaxis.set_label_position("right")
ax4.yaxis.tick_right()
ax5.yaxis.set_label_position("right")
ax5.yaxis.tick_right()
ax6.yaxis.set_label_position("right")
ax6.yaxis.tick_right()

ax0.axhline(y=0, c='k', zorder=-1, lw=1)
ax0.scatter(soap.Maxflux.values - soap.Minflux.values, soap.SOAPamp_0.values, alpha=0.05)
ax0.set_ylim(-0.001, 0.0007)
ax0.set_title("Degree 0")

ax2.axhline(y=0, c='k', zorder=-1, lw=1)
ax2.scatter(soap.Maxflux.values - soap.Minflux.values, soap.SOAPamp_2.values, alpha=0.05)
ax2.set_ylim(-0.0008, 0.001)
ax2.set_title("Degree 2")

ax3.axhline(y=0, c='k', zorder=-1, lw=1)
ax3.scatter(soap.Maxflux.values - soap.Minflux.values, soap.SOAPamp_3.values, alpha=0.05)
ax3.set_ylim(-0.001, 0.001)
ax3.set_title("Degree 3")

ax4.axhline(y=0, c='k', zorder=-1, lw=1)
ax4.scatter(soap.Maxflux.values - soap.Minflux.values, soap.SOAPamp_4.values, alpha=0.05)
ax4.set_ylim(-0.0005, 0.0005)
ax4.set_title("Degree 4")

ax5.axhline(y=0, c='k', zorder=-1, lw=1)
ax5.scatter(soap.Maxflux.values - soap.Minflux.values, soap.SOAPamp_5.values, alpha=0.05)
ax5.set_ylim(-0.0005, 0.0005)
ax5.set_title("Degree 5")

ax6.spines['top'].set_visible(False)
ax6.spines['right'].set_visible(False)
ax6.spines['left'].set_visible(False)
ax6.spines['bottom'].set_visible(False)
ax6.set_xticks([])
ax6.set_yticks([])


fig.add_subplot(111, frame_on = False)
plt.tick_params(labelcolor="none", bottom=False, left=False)
plt.xlabel("Line Depth", fontsize=14)
plt.ylabel("rPCA Loadings", fontsize=14)

plt.tight_layout()
plt.savefig("rpca_relativeamps.pdf", format='pdf')
plt.show()
```
